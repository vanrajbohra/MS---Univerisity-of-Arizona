# while

```python
# Print numbers from 1 to 25

i = 1 # Initialization
while i<=25: # condition
    print(i, end=" ")
    
    i = i+1 # increment/decrement
```

```python
# break

for i in range(50):
    if i==11:
        break
    else:
        print(i,end=" ")
```

```python
# continue

for i in range(50):
    if i==11:
        continue
    else:
        print(i,end=" ")
```

```python
# Factorial of a number

def facto(no): # 5
    f = 1
    for i in range(1,no+1):
        f = f*i
    return f
```

```python
# Average marks of a student
marks = [90,98,56,75,59]

def avg_marks(l):
    total=0
    for i in l:
        total += i
    return (total/(len(l)))

avg_marks(marks)
```

# Recursion

A function calling itself.

```python
def facto(no):
    if no==0:
        return 1
    else:
        return no*facto(no-1)
    
facto(5)
```

```python
k = int(input())
rooms = (int(x) for x in input().split(" "))

seen = {} # Set, as set can store only unique values

for i in rooms:
    if not i in seen:
        seen[i] = 1
        print("if:",seen)
    else:
        seen[i] += 1
        print("else:",seen)
        
for key, value in seen.items():
    if value != k:
        print(key)
```
