```python
import numpy as np
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")
```


### What is a *machine learning* model? 

In Layman terms:
A machine learning model is a way to relate the input features we have to the output we aim to find. 

###### Data 

Ficticious house pricing data. 

| Year_of_constuction      | Area (in sqft) | Price in USD|
| ----------- | ----------- |----------- |
|1950	|850	|119000 |
|2000	|750	|129000  |
|2018	|1000	|146000  |
|2020	|1289	|121000  |
|1980	|3564	|199000  |
|1975	|1744	|130000  |

An ML model should be able to relate the features 'year_of_construction' and 'Area' with the price of the house. 

```python
#define features
Year = np.array([1950,2000,2018,2020,1980,1975])
area = np.array([850,750,1000,1289,3564,1744])

#define prices 
real_price = np.array([119000,129000,146000,121000,199000,130000])
```

### 

```python
sorted_year         =np.sort(Year)  
sorted_area         =np.sort(area) 
sorted_real_price   =np.sort(real_price)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2)
fig.suptitle('Relation between features and Label')
ax1.scatter(sorted_year, sorted_real_price)
ax1.set_ylabel('House Price')
ax1.set_xlabel('Year')
ax2.scatter(sorted_area, sorted_real_price)
ax2.set_ylabel('House Price')
ax2.set_xlabel('Area')
plt.tight_layout()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2)
fig.suptitle('Relation between features and Label')
ax1.plot(sorted_year, sorted_real_price)
ax1.set_ylabel('House Price')
ax1.set_xlabel('Year')
ax2.plot(sorted_area, sorted_real_price)
ax2.set_ylabel('House Price')
ax2.set_xlabel('Area')
plt.tight_layout()
```
