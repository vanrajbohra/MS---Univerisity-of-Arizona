```python
df.drop(['Cabin'],axis = 1,inplace = True)
df.isna().sum().sort_values(ascending = False)
```

```python
sns.distplot(df.Age)
plt.axvline(df.Age.quantile(0),color = 'green')
plt.axvline(df.Age.quantile(0.25),color = 'green')
plt.axvline(df.Age.quantile(0.50),color = 'blue')
plt.axvline(df.Age.quantile(0.75),color = 'green')
plt.axvline(df.Age.quantile(1),color = 'green')
plt.axvline(df.Age.mean(),color = 'black')
plt.show()
```

- Age Data is normally distributed
- Mean and median are near to each other
- can substitue missing with mean or median


```python
df.Age.fillna(df.Age.median(),inplace = True)
df.isna().sum().sort_values(ascending = False)
```

```python
df.Embarked.unique()
```

```python
df.Embarked.value_counts()
```

- Most of the Passenger Embarked from SouthHampton
    - can be coz it is very near to London
    - can be coz its starting point

lets fill the missing value in this categorical column by the mode of the Embarked Column

```python
df.Embarked.fillna(df.Embarked.mode()[0],inplace = True)
df.isna().sum().sort_values(ascending = False)
```

## Data is Clean Now

## Feature Engineering

```python
#Create a Family Size Column

df['FamilySize'] = df.Parch + df.SibSp + 1
df.head()
```

```python
df.FamilySize.value_counts()
```

## Segment Sex Column as per Age, if Age<15 then Child else Sex Column Data must be copied



```python
df.sample(5)
```

```python
df["GenderClass"] = df.apply(lambda x : 'Child' if x.Age <15 else x.Sex,axis = 1)
```

```python
df.GenderClass.value_counts(normalize = True) * 100
```
